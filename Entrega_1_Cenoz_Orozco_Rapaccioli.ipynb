```python
#Runge Kutta Ej4 iii)

#defino la funcion de runge kutta a orden 4:
""" Calculo numericamente X(t + dt) a partir de X(t) y el campo de velocidades

 X: Da la solucion X(t) como un array de dim d.
 u: es el campo de velocidades como funcion de (x,t)
 t y dt: son respectivamente el tiempo y el incremento temporal.

 La funcion devuelve X(t+dt) como array de dim d."""

def runge4(X, u, t, dt):
    k1 = u(X, t)
    k2 = u(X + 0.5 * dt * k1, t + 0.5 * dt)
    k3 = u(X + 0.5 * dt * k2, t + 0.5 * dt)
    k4 = u(X + dt * k3, t + dt)

    Xn = X + (dt / 6) * (k1 + 2 * k2 + 2 * k3 + k4)
    return Xn
```

```python
#Trayectoria Ej4 iii)

#defino una funcio para la trayectoria

"""Aca calculo numericamente la solucion dx/ds = u(x,t) como un array usando la funcion runge 4 definida antes.

    x0:  es la ci c√≥mo array de dim d
    u: Campo de velocidades c√≥mo funci√≥n de (x,t)
    N: N√∫mero de pasos
    dt: Paso temporal
"""

def trayectoria(x0, u, N, dt):

    d = x0.shape[0]
    ts = np.arange(0, N+1)*dt
    xs = np.zeros((N+1, d))
    xs[0] = x0

    for j in range(N):
        xs[j+1] = runge4(xs[j], u, ts[j], dt) # Aplico RK4 sobre Xs[j] para obtener Xs[j+1]

    return ts, xs

Q,a=1,2

#Campo de velocidades
def U(X,t):
  return np.array([(Q/(X[0]**2 + X[1]**2))*X[0]+a*t, (Q/(X[0]**2 + X[1]**2))*X[1]])

N = 1000
dt = 1e-2
p0s = np.array([[-1, 2], [-1, 2], [-1 ,3], [1, 3], [2, 3], [2, 5]])

plt.figure(figsize=(8, 6))
for i in range(6):
  _, Xs = trayectoria(p0s[i], U, N, dt)
  plt.plot(Xs[0,0], Xs[0,1], "ko")
  plt.plot(Xs[:,0], Xs[:,1], "b-")
plt.grid(True)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('L√≠neas de trayectoria',fontsize=16)
```

```python
#Lineas de corriente Ej 4 iii)

#primero defino una funcion llamada lineasC:

""" Calcula numericamente la solucion como un array de dimension d

definiendo los parametros:
 - l0: es el origen de la linea como array de dim d
 - u: campo de velocidades (x,t)
 - t: tiempo
 - N: numero de pasos
 - ds: intervalo de los pasos
 """

def lineasC(l0, u, t, N, ds):

    d = l0.shape[0]
    s = np.arange(0, N+1)*ds
    ls = np.zeros((N+1, d))
    ls[0] = l0
    for j in range(N):
        ls[j+1] = runge4(ls[j], u, t, ds)

    return s, ls


Q, a = 1, 2
N = 10000
ds = 1e-3
p0s = np.array([[-1, 2], [-1, 2], [-1 ,3], [1, 3], [2, 3], [2, 5]])

plt.figure(figsize=(8, 6))
for i in range(6):
  _, ls = lineasC(p0s[i], U, 0, N, ds)
  plt.plot(ls[0,0], ls[0,1], "ko")
  plt.plot(ls[:,0], ls[:,1], "b-")
plt.grid(True)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('L√≠neas de corriente', fontsize=16)


```

## Problema 12 - Estrellas Autogravitantes
Este problema tiene como objetivo modelar la estructura b√°sica de una estrella, pensando que esta se compone de una masa de fluido auto-gravitande cuyos elementos se encuentran en equilibrio hidrost√°tico.
La presion evita que la estrella colapse por atraccion gravitaria. Adem√°s, suponemos simetr√≠a esf√©rica.
Para el fluido que compone la estrella, la ecuaci√≥n de Euler puede escribirse como:


$$
\frac{\vec\nabla p}{\rho}= - \frac{G M(r)}{r^2}\hat{r}
$$

con $\int_0^r \rho 4\pi r^2dr$ y $\hat{r}$ el versor radial en coordenadas esf√©ricas. Utilice la Ley de Gauss del campo gravitatorio $\nabla \cdot \vec g = -4\pi G \rho$ y que la fuerza gravitatoria por unidad de volumen es $\vec f_V=\rho \vec g$.
Numericamente quiero resolver los incisos $(v)$ y $(vi)$ de este problema. Tengo la Ec:

$$\frac{d^2 \theta}{d \xi^2} +\frac{2}{\xi} \frac{d\theta}{d\xi}+\theta^n=0 $$
 con las condiciones iniciales:

 $$  \theta (\xi=0)=1\\ \frac{d\theta}{d\xi}(\xi=0)=0$$


```python
#primero defino el sistema de ecs diferenciales de primer orden

def f(xi, t, dt, n):

    d2t = -2/xi * dt - t**n
    return dt, d2t

#defino Runge Kutta a 4orden:

def runge_kutta_4(xi, t, dt, n, h):

    K1_t, K1_dt = f(xi, t, dt, n)
    K2_t, K2_dt = f(xi + 0.5 * h, t + 0.5 * h * K1_t, dt + 0.5 * h * K1_dt, n)
    K3_t, K3_dt = f(xi + 0.5 * h, t + 0.5 * h * K2_t, dt + 0.5 * h * K2_dt, n)
    K4_t, K4_dt = f(xi + h, t + h * K3_t, dt + h * K3_dt, n)

    t_2 = t + (h / 6.0) * (K1_t + 2 * K2_t + 2 * K3_t + K4_t)
    dt_2 = dt + (h / 6.0) * (K1_dt + 2 * K2_dt + 2 * K3_dt + K4_dt)

    return t_2, dt_2
```

```python


#ahora loopeo para graficar:

for i in range(1,11):

    #Primero le doy condiciones iniciales para el problema
    xi_0 = 1e-16 #me evita la singularidad de dividir por cero.
    t_0 = 1
    dt_0 = 0
    n = i

    #doy los parametros de integracion:
    h = 0.01
    xi_f = 1

    #Creo listas donde guardar los resultados
    xi_l = [xi_0]
    t_l = [t_0]
    dt_l = [dt_0]

    #punto de partida
    xi = xi_0
    t = t_0
    dt = dt_0

    while xi < xi_f and t>=0:

        t, dt = runge_kutta_4(xi, t, dt, n, h)
        xi += h
        xi_l.append(xi)
        t_l.append(t)
        dt_l.append(dt)

    plt.plot(xi_l, t_l, label = f'n={n}')

    plt.xlabel(r'$\xi$',fontsize = 14)
    plt.ylabel(r'$\theta$',fontsize = 14)
    plt.title(r'Densidad de la estrella $ \theta (\xi = 1)$', fontsize = 16)
    plt.legend()

plt.grid(True)
plt.show()



```

```python
#Curva de la masa:


n_values = np.array([1, 2, 3, 4])
xi_final = 1000
h = 0.01

masas = []

for n in n_values:
    # Condiciones iniciales
    xi_0 = 1e-16  # Para evitar divisi√≥n por cero en 1/xi
    t_0 = 1
    dt_0 = 0

    # Listas para almacenar resultados
    xi_l = [xi_0]
    t_l = [t_0]
    dt_l = [dt_0]

    # Empiezo desde las CI
    xi = xi_0
    t = t_0
    dt = dt_0

    while xi < xi_final and t >= 0:
        t, dt = runge_kutta_4(xi, t, dt, n, h)
        xi += h
        xi_l.append(xi)
        t_l.append(t)
        dt_l.append(dt)

    # Calcular (xi^2) * dt_values * (-3)
    result = (np.array(xi_l)**2) * np.array(dt_l) * (-3)
    masas.append(result[-1])  # Tomar el √∫ltimo valor

# Graficar masas en funci√≥n de n
plt.plot(n_values, masas, marker='o', linestyle='-', color='g')
plt.title("Masa total adimensionalizada para n=1,2,3,4", fontsize = 16)
plt.xlabel('n', fontsize = 14)
plt.ylabel(r'$M_{T}/M_{0}$', fontsize =14)
plt.grid(True)
plt.show()
```
